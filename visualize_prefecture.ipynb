```python
# 必要なライブラリのインポート
import pandas as pd
import folium
from folium import plugins
```

```python
# 計算した感染者データ取得
infector_data = pd.read_csv("prefecture_infector_data.csv", encoding="utf-8", index_col=0)
```

```python
# 緯度経度データ取得
geo_df = pd.read_csv("prefecture_coordinate_data.csv", encoding="utf-8")
```

```python
# ヒートマップデータを作成
heatmap_data = [[[float(geo_df[geo_df["nam_ja"] == prefecture_name]["lat"]), float(geo_df[geo_df["nam_ja"] == prefecture_name]["lon"]),
                  row_data[day]] for prefecture_name, row_data in infector_data.iterrows()] for day in infector_data.columns]

# 日本地図を作成
japan_map = folium.Map(location=[35, 135], zoom_start=6)

# ヒートマップのインスタンスを生成する(円の半径や, ヒートマップの濃淡の設定ができる)
hm = plugins.HeatMapWithTime(heatmap_data,index=[day for day in infector_data.columns],auto_play=False,radius=20,max_opacity=1, gradient={0: 'blue', 0.03: 'lime', 0.13:'yellow',0.30: 'red'})

# ヒートマップのインスタンスに日本地図を追加する
hm.add_to(japan_map)

# ヒートマップの表示
japan_map
```
